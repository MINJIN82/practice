Numpy

```python
import numpy as np

a = np.arange(15).reshape(3, 5) # 3행 5열
print(a)
```

```python
a = np.array([[1,2],[3,4]])
 
s = np.sum(a)
print(s)  
 
# axis=0 이면, 컬럼끼리 더함
# axis=1 이면, 행끼리 더함
s = np.sum(a, axis=0)
print(s)   
 
s = np.sum(a, axis=1)
print(s)   
 
s = np.prod(a)  #prod = 전체 곱하기
print(s)  
```

```python
import pandas as pd


df = pd.read_csv("/content/drive/MyDrive/연습/감귤2.csv", encoding= "cp949") #encoding = "utf-8"6
df
```

```python
df = df.dropna(axis=1)  #nan값 삭제
```

```python
df = df.iloc[0:53,0:8] # 전일대비 등락 열까지 데이터 확인
```

```python
df
```

```python
df.columns = ['날짜',	'품목',	'등급',	'거래단위',	'최저가',	'최고가',	'평균가',	'전일대비등락'] #컬럼 이름 정리
```

```python
df1 = df.sort_values(by = ["날짜"], ascending = [True]) #오름차순, 데이터프레임 안에서 사용할땐 sort 대신 sort_values 사용만 가능
df1
```

```python
df_date_03 = df1[df1["날짜"] == "2023.01.03"] #1/3 데이터만 따로 정리
```
